## Objective
The objective of the dataset is to diagnostically predict whether or not a patient has diabetes, based on certain diagnostic measurements included in the dataset. Several constraints were placed on the selection of these instances from a larger database. In particular, all patients here are females at least 21 years old of Pima Indian heritage.

```python
data = pd.read_csv('./data/diabetes.csv')
data.head(10)
```

```python
data.info()
```

```python
data.describe()
```

```python
# remove zeros
data_copy = data.copy(deep = True)
data_copy[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']]=data[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']].replace(0,np.NaN)
```

```python
data_copy.head()
```

```python
print(data_copy.isnull().sum())
```

```python
data.hist(figsize = (20,20))
plt.show()
```

```python
# Fill na values with mean and median based on distribution
mean_list = ['Glucose','BloodPressure']
for i in mean_list:
    data_copy[i].fillna(data_copy[i].mean(),inplace = True)
```

```python
median_list = ['SkinThickness','Insulin','BMI']
for i in median_list:
    data_copy[i].fillna(data_copy[i].median(),inplace = True)
```

```python
data_copy.hist(figsize = (20,20))
plt.show()
```

```python
data_copy.shape
```

```python
data.Outcome.value_counts().plot(kind="bar")
```

```python
sns.pairplot(data_copy, hue = 'Outcome')
```



Pearson's Correlation Coefficient: helps you find out the relationship between two quantities. It gives you the measure of the strength of association between two variables. The value of Pearson's Correlation Coefficient can be between -1 to +1. 1 means that they are highly correlated and 0 means no correlation.

A heat map is a two-dimensional representation of information with the help of colors. Heat maps can help the user visualize simple or complex information.


```python
sns.heatmap(data_copy.corr(),annot=True,cmap='RdYlGn')
```

```python
X =  pd.DataFrame(data_copy.drop(["Outcome"],axis = 1),
        columns=['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction', 'Age'])
```

```python
X.head()
```

```python
y = data_copy.Outcome
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.33,random_state=42, stratify=y)
```

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler().fit(X_train)
X_train = sc_X.transform(X_train)
X_test = sc_X.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
testing_score,training_score = [],[]
for i in range(1,20):
    knn = KNeighborsClassifier(i)
    knn.fit(X_train,y_train)
    testing_score.append(knn.score(X_test,y_test))
    training_score.append(knn.score(X_train,y_train))
```

```python
plt.figure(figsize=(12,5))
p = sns.lineplot(range(1,20),training_score,marker='*',label='Train Score')
p = sns.lineplot(range(1,20),testing_score,marker='o',label='Test Score')
```

```python
max_testing_score = max(testing_score)
for index,val in enumerate(testing_score):
    if val == max_testing_score:
        print('Max test score is {}% and k is {}'.format(round(max_testing_score*100,2),index+1))
        break
```

```python
# final model:
knn = KNeighborsClassifier(11)
knn.fit(X_train,y_train)
knn.score(X_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
y_pred = knn.predict(X_test)
from sklearn import metrics
cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
p = sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="YlGnBu" ,fmt='g')
plt.title('Confusion matrix', y=1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
```

```python
confusion_matrix(y_test,y_pred)
```

```python
# classification report:
# Precision = TP/(TP+FP)
# Recall = TP/(TP+FN) {Fraction of +ve's correctly identified}
# F1 score: weighted avg of precision and recall. F1 Score = 2(Recall Precision) / (Recall + Precision)

from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_pred))
```

```python
from sklearn.metrics import roc_curve
y_pred_proba = knn.predict_proba(X_test)[:,1]
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
```

```python
plt.plot([0,1],[0,1],'k--')
plt.plot(fpr,tpr, label='Knn')
plt.xlabel('fpr OR 1-specificity')
plt.ylabel('tpr')
plt.title('Knn(n_neighbors=11) ROC curve')
plt.show()
```
